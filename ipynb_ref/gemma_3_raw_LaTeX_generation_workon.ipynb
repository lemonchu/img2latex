## Gemma 3 - 4B, 12B, 27B

```python
## FOR 4,12,27B model

from transformers import AutoProcessor, Gemma3ForConditionalGeneration
import torch

# model_id = "google/gemma-3-4b-it"
# model_id = "google/gemma-3-12b-it" # cell outputs below for from this model in nf4 (but I've defaulted a smaller model above)
model_id = "google/gemma-3-27b-it"

local_path = f"/mnt/moonfs/public-models-m2/{model_id}"

datatype = torch.bfloat16

# to use this, first run `!pip install bitsandbytes`
from transformers import BitsAndBytesConfig

# 4bit
quantization_config = BitsAndBytesConfig(
   load_in_4bit=True,
   bnb_4bit_quant_type="nf4",
   bnb_4bit_use_double_quant=True,
   bnb_4bit_compute_dtype=datatype
)

## 8bit
# quantization_config = BitsAndBytesConfig(
#    load_in_8bit=True,
# )

model = Gemma3ForConditionalGeneration.from_pretrained(
    # model_id,
    local_path,
    torch_dtype=datatype,
    device_map="auto",
    quantization_config=quantization_config # comment this out if you don't want to use quantization.
).eval()
```

```python
# processor = AutoProcessor.from_pretrained(model_id)
processor = AutoProcessor.from_pretrained(local_path)
```

```python
instruction="""
Convert the equation images to LaTeX equations.
"""

messages = [
    {
        "role": "system",
        "content": [{"type": "text", "text": "You are a helpful assistant."}]
    },
    {
        "role": "user",
        "content": [
            {"type": "image", "image": "img.jpeg"},
            {"type": "text", "text": instruction}
        ]
    }
]

inputs = processor.apply_chat_template(
    messages, add_generation_prompt=True, tokenize=True,
    return_dict=True, return_tensors="pt"
).to(model.device, dtype=torch.bfloat16)

input_len = inputs["input_ids"].shape[-1]

with torch.inference_mode():
    generation = model.generate(**inputs, max_new_tokens=512, do_sample=False)
    generation = generation[0][input_len:]

decoded = processor.decode(generation, skip_special_tokens=True)
print(decoded)
```
